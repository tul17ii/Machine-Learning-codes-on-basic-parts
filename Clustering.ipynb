```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# dp
X = np.array([[3, 11], [8, 4], [5, 8], [5, 11], [10, 7], [14, 9]])

# Initial cluster centers
init_centers = np.array([[3, 11], [5, 11], [20, 6]])

# KMeans with predefined centers (n_init=1 ensures only one init run)
kmeans = KMeans(n_clusters=3, init=init_centers, n_init=1, random_state=42)
kmeans.fit(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering")
plt.legend()
plt.show()

# Output
print("Cluster Centers:\n", kmeans.cluster_centers_)
print("Cluster Labels:", kmeans.labels_)
```
